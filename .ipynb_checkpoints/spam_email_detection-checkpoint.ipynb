<h3>3.Data Perprocessing</h3>
    <ul>
    <li>Lower Case</li>
    <li>Tokenization</li>
    <li>Removing Special Character</li>
    <li>Stemming</li>
    <li>Removing step words and punctuations</li>

</ul>

```python
import string
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
def data_process(t):
    t=t.lower()
    t=nltk.word_tokenize(t)
    l=[]
    for i in t:
        if i.isalnum():
            l.append(i)
    t=list(l)
    l.clear()
    for i in t:
        if i not in stopwords.words('english') and i not in string.punctuation:
            l.append(i)
    t=list(l)
    l.clear()
    for i in t:
        l.append(ps.stem(i))
    
    return " ".join(l)
```

```python
df['transformed_text']=df['item'].apply(data_process)
```

## 4. Model Building

```python
#applying bag of words for text vectorization
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
cv=CountVectorizer()
tfidf=TfidfVectorizer(max_features=1000)
X=tfidf.fit_transform(df['transformed_text']).toarray()
```

```python
y=df['type'].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,precision_score,confusion_matrix
```

```python
model=LogisticRegression(C=10.0)  # regularization parameter=10
#A higher value of C means less regularization (the model will fit the training data more closely
#while a lower value means more regularization (the model will generalize better to unseen data).
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
print(accuracy_score(y_test,y_pred))
print(precision_score(y_test,y_pred))
```

## Pipeline 

```python
from sklearn.pipeline import Pipeline
pipe= Pipeline([
    ('tfidf', TfidfVectorizer(max_features=1000)),
    ('logistic', LogisticRegression(C=10.0))
])

X1_train, X1_test, y1_train, y1_test = train_test_split(df['transformed_text'].astype(str),y, test_size=0.2, random_state=2)
X1_train
```

```python
pipe.fit(X1_train,y1_train)
```

```python
predict_value=pipe.predict(X1_test)
print(accuracy_score(y1_test,predict_value))
print(precision_score(y1_test,predict_value))
```

```python
email="You’ve won!: These emails claim that you have won a prize or lottery, but it’s a scam."
input_email=data_process(email) #Text preprocessing
pipe.predict([input_email])
```

```python
# ngram_range=(1,1), it means that only unigrams are considered. Unigrams are single words in the text.
# For example, let's consider the sentence: "The quick brown fox jumps over the lazy dog."
# output will be ["The", "quick", "brown", "fox", "jumps", "over", "the", "lazy", "dog"]

# max_df stands for maximum document frequency.


pipe1= Pipeline([
    ('tfidf', TfidfVectorizer(max_df=0.5,max_features=3000,ngram_range=(1,1))),
    ('MultinomialNB', MultinomialNB(alpha=0.1))
])
```

```python
pipe1.fit(X1_train,y1_train)
```

```python
predict_value1=pipe1.predict(X1_test)
print(accuracy_score(y1_test,predict_value1))
print(precision_score(y1_test,predict_value1))
```

## Hyperparameter tuning

```python
params={
    'tfidf__max_features': [1000, 2000, 3000],
    'logistic__C': [0.1, 1.0, 10.0]
}
```

```python
#  if cv=5, it means that the data will be divided into 5 folds, and the model will be trained and evaluated 5 times, 
#     each time using a different 80% of the data for training and 20% for validation.

from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(pipe, params, cv=5, n_jobs=-1, verbose=1)
grid_search.fit(X1_train, y1_train)
```

```python
print("Best Parameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)
```
